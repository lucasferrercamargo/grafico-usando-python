```python
brasi_dict = {'ano' : brasil.index.tolist(), 'imigrantes' : brasil.values.tolist()}
dados = pd.DataFrame(brasi_dict)
```

```python
from matplotlib import colors
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados['ano'], dados['imigrantes'], lw=3, color='g')
ax.set_title('Numero de imigrantes do Brasil para o canadá', fontsize=18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Numero de imigrantes', fontsize=14),
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)



plt.show()
```
